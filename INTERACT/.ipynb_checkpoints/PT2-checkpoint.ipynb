## Manning's Equation

```python
y = 0.5
n = 0.014
S = 0.00125
c = 0.3
W = 4.5
# A = W*y - c*c
# P = W - 2*c + 2*(y-c) + 2*c*np.sqrt(2)
# R = A/P
# V = 1/n*R**(2/3)*S**(1/2)
# Q = V*A
```

```python
def A(y):
  return W*y - c*c
```

```python
def P(y):
  return W - 2*c + 2*(y-c) + 2*c*np.sqrt(2)
```

```python
def R(y):
  return A(y)/P(y)
```

```python
def V(y):
  return 1/n*(R(y))**(2.0/3.0)*S**(1.0/2.0)
```

```python
def Q(y):
  return V(y)*A(y)
```

```python
interact(Q,y=(0.3,5,0.1));
```
